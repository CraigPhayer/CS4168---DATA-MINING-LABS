## How does age or job correlate with subscription rates?

```python
df['age'].mean()
```

```python
df['balance'].median()
```

```python
df['duration'].std()
```

```python
df['marital'].value_counts(normalize=True)
```

```python
# age distribution
df['age'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Age vs Subscription rate
sns.boxplot(x='subscribed', y='age', data=df)
plt.title('Age vs Subscription rate')
plt.show()
```

### Age distribution
- Age distribution is right-skewed with most people in their 30s and 40s with a mean age of 41.75 and the number of people decreasing as age increases.
- The banks marketing campaign is more successful with younger people as the subscription rate is higher for people in their 20s and 30s.

```python
# job distribution
df['job'].value_counts()
```

```python
df['job'].value_counts().plot(kind='bar')
plt.xlabel('Job')
plt.ylabel('Frequency')
plt.show()

# Job vs Subscription rate
sns.countplot(x='job', hue='subscribed', data=df)
plt.xticks(rotation=45)
plt.title('Job vs Subscription rate')
plt.show()
```

```python
# Age and Balance by Subscription rate
sns.scatterplot(x='age', y='balance', hue='subscribed', data=df)
plt.title('Age and Balance by Subscription rate')
plt.show()

# Marital Status and Education by Subscription rate
sns.countplot(x='marital', hue='subscribed', data=df)
plt.title('Marital Status and Education by Subscription rate')
plt.show()
```

```python
encoded_data = pd.get_dummies(df, drop_first=True)

sns.heatmap(encoded_data.corr(), annot=False, cmap='coolwarm');
```
