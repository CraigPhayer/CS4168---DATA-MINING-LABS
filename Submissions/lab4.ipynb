```python
# imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc
from sklearn.neural_network import MLPClassifier

import warnings

warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('../Data/seeds.csv')
df.head(10)
```

```python
df.tail(10)
```

```python
df.isna().sum()
```

the plot below shows the `type` column, which represents the original three seed classes.

```python
df.hist(bins=20, figsize=(20, 15), color='b')
```

```python
df['type'].value_counts()
```

To perform a binary classification, we will convert the `type` column to a binary class. We will choose type `1` as class `0` and merge the remaining two types (type `2` and `3`) into class `1`.

```python
# Convert the type column to a binary class
df['class'] = df['type'].apply(lambda x: 0 if x == 1 else 1)
df = df.drop(columns=['type'])

df['class'].value_counts()
```

Prep the data for training by splitting into training and testing.

```python
x = df.drop(columns=['class'])
y = df['class']

# Split into 80% train 20% test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Pipelines

SVM_pipline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', SVC(probability=True, random_state=42, kernel='linear'))
])

RF_pipline = Pipeline([
    ('model', RandomForestClassifier(n_estimators=100, random_state=42))
])

MLP_pipline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', MLPClassifier(hidden_layer_sizes=(50,),random_state=42, max_iter=1000))
])
```

```python
# Train

SVM_pipline.fit(X_train, y_train)
```

```python
RF_pipline.fit(X_train, y_train)
```

```python
MLP_pipline.fit(X_train, y_train)
```

The classification report prints precision, recall, f1-score, and accuracy for each model, helping us to compare the performance of the models and even shows how well each model handled the 2 individual classes.

```python
y_pred_SVM = SVM_pipline.predict(X_test)
print("SVM Classification Report:\n", classification_report(y_test, y_pred_SVM))

y_pred_RF = RF_pipline.predict(X_test)
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_RF))

y_pred_MLP = MLP_pipline.predict(X_test)
print("MLP Classification Report:\n", classification_report(y_test, y_pred_MLP))
```

## Classification Report Summary

- **SVM**:
    - SVM achieved a balanced accuracy of 90%, with similar performance on both classes.
- **Random Forest**:
    - Random Forest had perfect precision on class 0 but a lower recall when compared to the other two models at .64 recall, suggesting it was more confident but missed some positives.
- **MLP**:
    - MLP also achieved 90% accuracy, with a slightly better balance between classes than Random Forest.

## ROC Curve

```python
y_prob_SVM = SVM_pipline.predict_proba(X_test)[:,1]
y_prob_RF = RF_pipline.predict_proba(X_test)[:,1]
y_prob_MLP = MLP_pipline.predict_proba(X_test)[:,1]

# ROC curve
fpr_SVM, tpr_SVM, _ = roc_curve(y_test, y_prob_SVM)
fpr_RF, tpr_RF, _ = roc_curve(y_test, y_prob_RF)
fpr_MLP, tpr_MLP, _ = roc_curve(y_test, y_prob_MLP)

plt.plot(fpr_SVM, tpr_SVM, label='SVM')
plt.plot(fpr_RF, tpr_RF, label='Random Forest')
plt.plot(fpr_MLP, tpr_MLP, label='MLP')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

### ROC Results

To better visualize the trade-off between true positives and false positives, we plot the ROC curves for each model.

- All models show a strong performance.
- The **MLP classifier** achieved nearly perfect AUC, with a curve closely following the top-left corner.
- **SVM** followed closely behind with a slightly lower AUC but still resembles a very similar curve to the **MLP**.
- The **Random Forest** curve starts the same as the two other models, but then levels off, suggesting it is less confident in the middle range of the curve, again showing where it might have missed some positives.
